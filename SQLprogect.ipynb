**5. Исследование данных с помощью SQL**

**Wich films have highest rates?**

Я выбрала 20 фильмов с самым высоким рейтигом

```python
query = '''SELECT  NAME, DATE, RATE, GENRE
           FROM Imbd
           WHERE RATE <> "No Rate" AND TYPE = "Film"
           ORDER BY RATE DESC

           LIMIT 20	'''
```

```python
for row in connection.execute(query):
      print(row)
```

**Wich films have highest numbers of votes?**

Я выбрала 20 фильмов с самым большим количеством голосов

```python
query = '''SELECT  NAME, DATE, VOTE
           FROM Imbd
           WHERE VOTE <> "No Votes" AND TYPE = "Film"

           ORDER BY VOTE DESC

           LIMIT 20	'''
```

```python
for row in connection.execute(query):
      print(row)
```

**How genre affects rate ?**

```python
query = '''SELECT
           AVG(RATE),
           GENRE
           FROM Imbd
           GROUP BY GENRE
           ORDER BY AVG(RATE) DESC

           LIMIT 10	'''
```

```python
for row in connection.execute(query):
      print(row)
```

Здесь видно, что наибольший рейтинг имеют семейные фильмы, комедии и анимационные фильмы

**Does series get higher rates ?**

```python
query = '''SELECT TYPE, AVG(RATE)
           FROM Imbd
           GROUP BY TYPE '''
```

```python
for row in connection.execute(query):
      print(row)
```

Мы видим, что не на много, но да.
